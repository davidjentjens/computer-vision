# Comparação entre filtros FTT e Convoução

Abaixo podemos ver as diferenças entre os dois tipos de filtros.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('moonlanding_ftt.png')
fig = plt.imshow(img)
plt.axis('off')
plt.rcParams["figure.figsize"] = (16,10)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
```

```python
img = mpimg.imread('moonlanding_convolution.png')
fig = plt.imshow(img)
plt.axis('off')
plt.rcParams["figure.figsize"] = (16,16)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
```

Podemos observar que o filtro por convolução suaviza o ruído, enquanto o filtro pela transformada de fourier chega a intensificar um pouco da cor a imagem e adicionar um blur, para chegar mais perto de remover o ruído.
